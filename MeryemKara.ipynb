```python
import pandas as pd
import zipfile

# Unzip the archive to access the actual data file
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    # List all files inside the zip archive to identify the correct path
    print(zip_ref.namelist())
    zip_ref.extractall('.')

# Now, load the correct data file
df = pd.read_csv("Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")
print(df.head())
print(df.info())
```

```python
print(df['gender'].value_counts())

print(df['age'].value_counts().sort_index())
```

```python
from google.colab import files

# DataFrame'i CSV dosyasına kaydet
df.to_csv('islenmis_veri.csv', index=False)

# Dosyayı indir
files.download('islenmis_veri.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='gender')
plt.title("Kullanıcı Cinsiyet Dağılımı")
plt.show()

plt.figure(figsize=(12,6))
df['age'].value_counts().sort_index().plot(kind='line')
plt.title("Yaşa Göre Kullanıcı Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Kullanıcı Sayısı")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['stress_level'].value_counts().head(10).plot(kind='bar')
plt.title("Kullanıcı Stres Seviyesi Dağılımı")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['addiction_level'].value_counts().plot(kind='bar')
plt.title("Kullanıcı Bağımlılık Seviyesi Dağılımı")
plt.show()
```
